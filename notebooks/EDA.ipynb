```python
import numpy as np

# Calculate the distance between the restaurant and the delivery location using the Haversine formula
R = 6371  # Radius of the Earth in kilometers
lat1, lon1 = np.radians(df[['Restaurant_latitude', 'Restaurant_longitude']].values.T)
lat2, lon2 = np.radians(df[['Delivery_location_latitude', 'Delivery_location_longitude']].values.T)

delta_lat = lat2 - lat1
delta_lon = lon2 - lon1

a = np.sin(delta_lat/2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(delta_lon/2) ** 2
c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

distance = R * c  # distance in kilometers

# Add the distance column to the dataframe
df['distance'] = distance
```

```python
df = df.drop(["Restaurant_latitude", "Restaurant_longitude", "Delivery_location_latitude", "Delivery_location_longitude", "Order_datetime","picked_datetime"], axis = 1)
```

```python
q1 = df["distance"].quantile(0.25)
q3 = df["distance"].quantile(0.75)
iqr = q3 - q1
upper = q3 + 1.5*iqr
lower = q1 - 1.5*iqr
outliers = df[(df["distance"] > upper) | (df["distance"] < lower)].index
print("Range:", upper , lower)
```
